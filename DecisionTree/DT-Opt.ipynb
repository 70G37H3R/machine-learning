## import libraries

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

### prepare data


```python
df_train = pd.read_csv("data/optics/opt_train.csv", delimiter=',', header=None, names=[i for i in range(65)], index_col=False)
```

```python
df_test = pd.read_csv("data/optics/opt_test.csv", delimiter=',', header=None, names=[i for i in range(65)], index_col=False)
```

### exploratory

```python
df_train.head(5)
```

```python
df_test.head(5)
```

### get features

```python
x_train = df_train.iloc[:,:-1]
x_train.head(5)
```

```python
x_test = df_test.iloc[:,:-1]
x_test.head(5)
```

### get labels

```python
y_train = df_train.iloc[:,-1:]
y_train.head(5)
```

```python
y_test = df_test.iloc[:,-1:]
y_test.head(5)
```

### Normalize data

#### min max scale

```python
x_train = (x_train-x_train.min())/(x_train.max()-x_train.min())
x_train.head(5)
```

```python
x_train.fillna(0, inplace = True)
x_train.head(5)
```

```python
x_test = (x_test-x_test.min())/(x_test.max()-x_test.min())
x_test.head(5)
```

```python
x_test.fillna(0, inplace = True)
x_test.head(5)
```

### Trainning

```python
classifier = DecisionTreeClassifier()
classifier.fit(x_train, y_train.values.ravel())
```

```python
y_pred = classifier.predict(x_test) 
```

### Evaluate

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
